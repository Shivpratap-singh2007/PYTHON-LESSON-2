## <u> Slicing </u>
***<u>Accessing parts of a string</u>***

str[strting_idx:ending_idx] #ending idx is not included 

str = "ShivPratap"    
str[1 : 4] is "hiv"   
str[ : 4] is same as str[0 : 4]     
str[1 : ] is same as str[1 : len(str)]     

```python
str = "Shivpratap"
print(str[1:5])
print(str[1 : len(str)])
print(str[1:])
print(str[:6])
```
